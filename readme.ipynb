## Purpose

This project takes a DEM raster and a polygon (User created) and analyzes them to show the areas in the polygon that receive full sun, partial sun, or shade. It allows for anyone from a home gardener trying to determine plotting areas to a scientist trying to determine the sun exposure on a study site. Other possible applications include builders determining sun exposure. 

## Data Inputs

This tool requires a feature class polygon that overlaps with the DEM raster and a DEM raster in .tif format. The polygon must overlap with an area with sufficient elevation data. 

## Outputs

This tool produces:
- an output geodatabase if one does not already exist
- a classification of the sun exposure (1, 2, 3). The classification is broken down inside of the attribute table of the reclassification
- a PNG file of the polygon with a legend

Outputs are saved in an output folder that the user chooses. Look for all of the commands completing in the dialogue box as the tool runss, and then make sure that the PNG file was created and exported into the folder. 

## Other information

This code needs Arcpy, matplotlib and numpy to run

### Common errors
This code may fail to execute if the polygon is too small or not overlapping the DEM correctly
